4.Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference.

```python
#As datatypes are object that's why first we have to change datatype to 'int' and then we use Goal_diff_count = GoalsFor - GoalsAgainst
def Goal_diff_count():
    df['Goal_diff_count'] = df['GoalsFor'].astype(int)-df['GoalsAgainst'].astype(int)
    return df[['Team','Goal_diff_count']].sort_values(by = 'Goal_diff_count',ascending=False)

Goal_diff_count()
        
```

 find the team which has maximum and minimum goal difference.

```python
print("The team which has maximum goal difference is :")
Goal_diff_count().head(1)
```

```python
print("The team which has maximum goal difference is :")
Goal_diff_count().tail(1)
```

5.Create a new column with name “Winning Percent” and append it to the data set.If there are any numerical error, replace it with 0%, and Print the top 5 teams which has the highest Winning percentage.

```python
#As datatypes are object that's why first we have to change datatype to 'int' and then we use Percentage of Winning = (GamesWon / GamesPlayed)*100
Percentage_of_Winning = (df['GamesWon'].astype(int) /df['GamesPlayed'].astype(int))*100
```

```python
#Appending to the dataset
df['Winning Percent'] = Percentage_of_Winning
```

```python
#Replacing nan with 0
df['Winning Percent'].fillna(0,inplace = True)
```

```python
df[['Team','Winning Percent']].head()
```

6.Group teams based on their “Best position” and print the sum of their points for all positions.

```python
df.groupby(['BestPosition'])['Points'].sum().sort_values(ascending = False).reset_index()
```
